```python
import pandas as pd
import requests
import json
import sqlite3
import sqlalchemy
```

```python
url = 'https://rickandmortyapi.com/api/'
```

```python
endpoint = 'character'
```

```python
r = requests.get(url + endpoint)
#r.json()
pretty_json = json.loads(r.text)
print(json.dumps(pretty_json, indent=4))
```

```python
def main(url, endpoint, x):
    r = requests.get(url + endpoint + f'?page={x}')
    return r.json()

def num_pages(response):
    return response['info']['pages']

def parse_json(response): 
    character_list = []
    for record in response['results']: 
        character = {
            'id': record['id'],
            'image': record['image'],
            'gender': record['gender'],
            'name': record['name'], 
            'episode': len(record['episode'])
        }
        character_list.append(character)
    return character_list
```

```python
ricky_morty_character =[]
data = main(url, endpoint, 1)
for page in range(1, num_pages(data)+1):
    print(f'ingesting data on page: {page}')
    ricky_morty_character.extend(parse_json(main(url, endpoint, page)))
```
